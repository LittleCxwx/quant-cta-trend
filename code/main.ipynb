# 获取筛选后每个股票从2000年至2025年12月31日前的交易日

```python
import akshare as ak
import pandas as pd
from tqdm import tqdm   # 进度条，pip install tqdm

# 1. 读代码表，强制用字符串，保留前导 0
code_df = pd.read_csv("..\data\stock_zh_a_spot_em_filtered.csv", dtype={"代码": str})

# 2. 循环拉日线
for code in tqdm(code_df["代码"], desc="下载日线"):
    # AKShare 东财接口要求深市加“sz”，沪市加“sh”
    symbol = code
    
    # 拉后复权日线（2000 至今）
    df = ak.stock_zh_a_hist(symbol=symbol,
                            period="daily",
                            start_date="20000101",
                            end_date="20251231",
                            adjust="")
    
    # 保存为“002133.csv”
    df.to_csv(f"../data/stock_daily_data/{code}.csv", index=False, encoding="utf-8-sig")

print("全部下载完成！")
```

```python
# 拼接一个完整的csv
import pandas as pd
from pathlib import Path

folder = Path(r"../data/stock_daily_data")   # 例如 r"D:\data\daily"
out_file = "../data/all_stocks_daily.csv"

# 1. 收集所有 csv 文件
csv_files = list(folder.glob("*.csv"))

# 2. 逐个读取并追加
chunks = []
for fp in csv_files:
    code = fp.stem                          # '002133'
    df = pd.read_csv(fp, dtype={'股票代码': str})   # 若文件里本身就有代码列可省略 dtype
    df.insert(0, '代码', code)              # 确保第一列是 str 类型的代码
    chunks.append(df)

# 3. 合并 + 排序
big = pd.concat(chunks, ignore_index=True)
big['日期'] = pd.to_datetime(big['日期'])
big = big.sort_values(['日期', '股票代码']).reset_index(drop=True)

# 4. 保存
big.to_csv(out_file, index=False, encoding='utf-8-sig')
print(f'已完成！共 {len(big)} 行，输出文件：{out_file}')
```

# 主要指标

```python
import akshare as ak
import pandas as pd
from tqdm import tqdm
```

```python
stock_list = pd.read_csv("../data/stock_zh_a_spot_em_filtered.csv", dtype={"代码": str})

def add_suffix(code: str) -> str:
    if code.startswith(("0", "2", "3")):
        return code + ".SZ"
    elif code.startswith("6"):
        return code + ".SH"
    elif code.startswith(("8", "9")):
        return code + ".BJ"
    else:
        return code

stock_list["代码_full"] = stock_list["代码"].str.zfill(6).apply(add_suffix)

# 保存为csv文件
stock_list.to_csv("../data/stock_zh_a_spot_em_filtered_with_suffix.csv", index=False)
```

```python
def load_annual_report(symbol_full: str):
    df = ak.stock_financial_analysis_indicator_em(
        symbol=symbol_full,
        indicator="按报告期"
    )

    # 只要年报
    df = df[df["REPORT_TYPE"] == "年报"].copy()

    # 时间
    df["NOTICE_DATE"] = pd.to_datetime(df["NOTICE_DATE"])
    df = df.sort_values("NOTICE_DATE")

    # ★ 只保留必要列 ★
    keep_cols = [
        "NOTICE_DATE",
        "EPSJB",
        "ROEJQ"
    ]
    df = df[keep_cols]

    return df

def load_market_value(symbol: str):
    df = ak.stock_value_em(symbol=symbol)
    df = df.rename(columns={"数据日期": "日期"})
    df["日期"] = pd.to_datetime(df["日期"])
    df = df.sort_values("日期")
    return df[["日期", "流通市值"]]

def process_single_stock(
    daily_df: pd.DataFrame,
    code: str,
    code_full: str,
    name: str
):
    df = daily_df[daily_df["股票代码"] == code].copy()
    if df.empty:
        return None

    df["日期"] = pd.to_datetime(df["日期"])
    df = df.sort_values("日期")

    # ====== 年报 ======
    try:
        annual_df = load_annual_report(code_full)
        df = pd.merge_asof(
            df,
            annual_df,
            left_on="日期",
            right_on="NOTICE_DATE",
            direction="backward"
        )
    except Exception as e:
        print(f"[年报失败] {code} {e}")

    # ====== 流通市值 ======
    try:
        mv_df = load_market_value(code)
        df = pd.merge_asof(
            df,
            mv_df,
            on="日期",
            direction="backward"
        )
    except Exception as e:
        print(f"[市值失败] {code} {e}")

    # ====== 补名称 ======
    df["名称"] = name
    df["股票代码_full"] = code_full

    return df

# 日行情 CSV（你最开始那个 2000–2025 的文件）
daily_df = pd.read_csv(
    "../data/all_stocks_daily.csv",
    dtype={"股票代码": str}
)

daily_df["日期"] = pd.to_datetime(daily_df["日期"])

# 只保留 2018-01-01 之后
daily_df = daily_df[daily_df["日期"] >= "2018-01-01"].copy()

result_list = []

for _, row in tqdm(stock_list.iterrows(), total=len(stock_list)):
    code = row["代码"].zfill(6)
    code_full = row["代码_full"]
    name = row["名称"]

    df_one = process_single_stock(
        daily_df=daily_df,
        code=code,
        code_full=code_full,
        name=name
    )

    if df_one is not None:
        result_list.append(df_one)

final_df = pd.concat(result_list, ignore_index=True)

final_df.to_csv(
    "../data/A股_日行情_年报_流通市值.csv",
    index=False,
    encoding="utf-8"
)
```

```python
import akshare as ak
import pandas as pd
def load_market_value(symbol: str, trade_date: pd.Timestamp):
    df = ak.stock_value_em(symbol=symbol)
    df["数据日期"] = pd.to_datetime(df["数据日期"])
    df = df[df["数据日期"] <= trade_date]

    if df.empty:
        return None

    return df.sort_values("数据日期").iloc[-1]["流通市值"]

df = load_market_value("000001", pd.Timestamp("2023-12-31"))
df
```
